```python
import requests
from bs4 import BeautifulSoup
import re

# very difficult trail
# URL = "https://www.mtbproject.com/trail/4932230/cathedral-rock-big-loop"
# URL = "https://www.mtbproject.com/trail/4915937/twin-buttes-loop"
URL = "https://www.mtbproject.com/trail/4939737/hangover-loop"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

# titlebar
titlebar = soup.find(id="title-bar")
# print(titlebar.prettify())

# header
trailTitle = soup.find(id="trail-title")
trailTitle = trailTitle.text.strip()
print("Trail title = " + trailTitle)

# split the trail name so we can set the id
trailTokens = trailTitle.split()
trail_id = trailTokens[0].lower()
print("Trail id = " + trail_id)

# difficulty
diffbanner = soup.find("div", class_="title")
difficulty = soup.find("span", class_="difficulty-text")
difficulty = difficulty.text.strip()
print("Difficulty = " + difficulty)

# trail subheader containing reviews
metaWrapper = soup.find("div", class_="stars-container")
topRatings = metaWrapper.find("span", class_="small")
totalRatings = topRatings.text.strip()
# print("topRatings = " + str(totalRatings))

# separate trail ratings and num ratings
ratingTokens = totalRatings.split()
numRatings = ratingTokens[1].replace('(', '')
numRatings = numRatings.replace(')', '')
avgRating = ratingTokens[0]

print("Number of ratings = " + numRatings)
print("Average rating = " + avgRating)
print("\n")

# Let's now create the main table for the trailz route
mtb_trail_route = {
    "_id": trail_id,
    "route_name": trailTitle,
    "difficulty": difficulty,
    "average_rating": avgRating,
    "num_ratings": numRatings
}
print("MTB trail route = ")
print(mtb_trail_route)

# TODO: We will set the bread crumbs at a later point
# breadcrumbs for the area of interest
# NOTE: THis contains a list of items, each a sublink to smaller areas
# ie: Areas -> AZ -> Flagstaff and Northern AZ -> Sedona
# crumbs = soup.find("ol", class_="breadcrumb")
# print("\n---Bread---")
# print(crumbs)
# print("\n")

# TODO: Trail summary will be implemented at a later time
# # trail quotes
# trailSummary = soup.find(id="trail-summary")
# trailQuote = trailSummary.find("h2", class_="text-xs-center")
# trailQuote = trailQuote.text.strip()
# print("--- Trail Summary ---")
# print(trailQuote)
# print("\n")
```

```python
# let's connect to our mongoDB instance and start storing stuff
from pymongo import MongoClient
from dotenv import dotenv_values

# let's create a basic function for getting the db
def get_database():
    config = dotenv_values(".env")
    URL_STRING = config["ATLAS_URI"]
    DB_NAME = config["DB_NAME"]

    # connect to the MongoDB client
    client = MongoClient(URL_STRING)
    mtb_db = client[DB_NAME]
    return mtb_db
```

```python
from pandas import DataFrame
# get the mtb db and query the collections
mtb_db = get_database()

trail_routes = mtb_db["mtb_trail_routes"]
trail_descriptions = mtb_db["mtb_trail_route_descriptions"]

# query the route info
routeItems = trail_routes.find()
descItems = trail_descriptions.find()

# convert the collections to data frames
routeDF = DataFrame(routeItems)
descDF = DataFrame(descItems)

print("--- Route DF ---")
print(routeDF)
print("\n")

print("--- Desc DF ---")
print(descDF)
print("\n")
```
