```python
import re

# create a function that takes in main text and parses
def parse_main_text(mainText):
    # NOTE: Main section text starts at index = 2
    start = 2
    end = len(mainText)
    count = 0
    mainBody = []
    for idx in range(start, end-1):
        elem = mainText[idx]
        strippedText = elem.text.strip()
        newText = re.sub(' +', ' ', strippedText)
        newText = re.sub('\n', '\n', newText)
        print("TEXT LEN = " + str(len(newText)))
        print("COUNT = " + str(count))
        print(newText)
        print("-----------------\n")
        mainBody.append(newText)
        count+=1
    
    return mainBody
        
    # TODO: Need to return the parsed text as array
```

```python
import requests
from bs4 import BeautifulSoup
import re

# very difficult trail
# URL = "https://www.mtbproject.com/trail/4932230/cathedral-rock-big-loop"
# URL = "https://www.mtbproject.com/trail/4915937/twin-buttes-loop"
URL = "https://www.mtbproject.com/trail/4939737/hangover-loop"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

# difficult trail
# URL = "https://www.mtbproject.com/trail/285392/tunnel-otero-west-figure-8"
# URL = "https://www.mtbproject.com/trail/4915937/twin-buttes-loop"
# page = requests.get(URL)
# soup = BeautifulSoup(page.content, "html.parser")

# # intermediate
# URL = "https://www.mtbproject.com/trail/7024744/figure-eight-mile-olio"
# page = requests.get(URL)
# soup = BeautifulSoup(page.content, "html.parser")

# # inter/difficult
# URL = "https://www.mtbproject.com/trail/912234/king-of-the-mountain-golden-eagle-loop"
# page = requests.get(URL)
# soup = BeautifulSoup(page.content, "html.parser")

# # easy/inter
# URL = "https://www.mtbproject.com/trail/193465/corrales-bosque"
# page = requests.get(URL)
# soup = BeautifulSoup(page.content, "html.parser")

# # easy
# URL = "https://www.mtbproject.com/trail/7013681/paseo-del-bosque-trail-loop"
# page = requests.get(URL)
# soup = BeautifulSoup(page.content, "html.parser")

# titlebar
titlebar = soup.find(id="title-bar")
# print(titlebar.prettify())

# header
trailTitle = soup.find(id="trail-title")
trailTitle = trailTitle.text.strip()
print("Trail title = " + trailTitle)

# difficulty
diffbanner = soup.find("div", class_="title")
difficulty = soup.find("span", class_="difficulty-text")
difficulty = difficulty.text.strip()
print("Difficulty = " + difficulty)

# trail subheader containing reviews
metaWrapper = soup.find("div", class_="stars-container")
topRatings = metaWrapper.find("span", class_="small")
totalRatings = topRatings.text.strip()
print("topRatings = " + str(topRatings))

# breadcrumbs for the area of interest
# NOTE: THis contains a list of items, each a sublink to smaller areas
# ie: Areas -> AZ -> Flagstaff and Northern AZ -> Sedona
crumbs = soup.find("ol", class_="breadcrumb")
print("\n---Bread---")
print(crumbs)
print("\n")

# trail quotes
trailSummary = soup.find(id="trail-summary")
trailQuote = trailSummary.find("h2", class_="text-xs-center")
trailQuote = trailQuote.text.strip()
print("--- Trail Summary ---")
print(trailQuote)
print("\n")

# TODO: Choose the trail API for the maps (https://openmaptiles.org/)
# and (https://developers.arcgis.com/python/)
# get the actual map of the trail route
# trailMap = soup.find(id="map-and-ride-finder-container")
# print(trailMap)
# print("\n")

# vertProfile = soup.find(id="vertical-profile")
# print(vertProfile)

# trail stats beneath map and gradient this contains a list of blocks
trailStats = soup.find_all("div", class_="stat-block")
print("--- Trail Stats ---")
for block in trailStats:
    print(block)
    print("\n")
    
# get the toolbox containing
# SECTIONS: driving directions, GPX, and image uploads
toolBox = soup.find(id="toolbox")
print("--- TOOLBOX ---")
print(toolBox)
print("\n")

# main content area
trailText = soup.find(id="trail-text")

#get the FEATURES
features = trailText.find_all("h3", class_="mr-2")
#     lambda tag: tag.name="h3" and text in 
#     "h3", string=lambda text: "features" in text.lower())
print("--- H3 Features ---")
print(features)
# for feature in features:
#     print(feature.text.strip())
print("\n")

# let's try and find the main text header sections
h3Tags = trailText.find_all("h3")
print("----- H3 TAGS -----")
print("H3 tags len = " + str(len(h3Tags)))
h3Count = 0
mainSectionHeaders = []
for tag in h3Tags:
    h3Class = tag.get('class')
    h3Text = tag.text.strip()
    h3Count += 1
    if h3Class == None:
        sectionHeader = re.sub(' +', ' ', h3Text)
        mainSectionHeaders.append(sectionHeader)

print("----- MAIN HEADERS -----")
print(mainSectionHeaders)
print("\n")


# print("--- TRAIL TEXT ---")
# print(trailText)
# print("\n")
# to parse text we need to remove things 

# TODO: Let's pull all of the text describing the trail, and we'll worry
# about the headers for Overview/Description for a later time
#     newText = " ".join(strippedText.strip().split())
#     newText = " ".join(strippedText.split())
mainText = trailText.find_all("div", class_="mb-1")
bodyText = parse_main_text(mainText)

# Let's create the dictionary of section headers to main text
mainTextMap = {}
start = 0
end = len(mainSectionHeaders)-1
for i in range(start, end):
    header = mainSectionHeaders[i]
    text = bodyText[i]
    mainTextMap[header] = text

print("Main map size = " + str(len(mainTextMap)))
for key in mainTextMap:
    print('{0} : {1}'.format(key, mainTextMap[key]))
```
